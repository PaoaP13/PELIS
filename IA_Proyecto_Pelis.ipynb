# PRE PROCESAMIENTO

```python
df1=pd.read_csv('tmdb_5000_movies.csv') #carga de archivos csv
df2=pd.read_csv('tmdb_5000_credits.csv')
```

```python
data=df1.merge(df2,on='title') # combina dataframes con 'title' y almacena el resultado en 'data'
data.head()
```

```python
data=data[['id','title','overview','genres','keywords','cast','crew']] # selecciona columnas específicas de 'df1'
data.head()
```

```python
data.isnull().sum() #número de valores nulos en cada columna
```

```python
data.dropna(inplace=True) #elimina filas que contienen valor nulo
data.shape #(filas, columnas)
```

```python
data.duplicated().sum() #número de filas duplicadas
```

```python
data.drop_duplicates(inplace=True) # eliminar filas duplicadas
```

```python
import ast #analiza y evalua expresiones/estructuras de datos literales
#Convertir cadena de objeto en una lista de diccionarios
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
data['genres']=data['genres'].apply(convert) # transforma datos de 'geners' a lista de nombres
data['keywords']=data['keywords'].apply(convert) # transforma datos de 'keywords' a lista de nombres
data.head()
```

```python
def conv(obj):
    c=0
    l=[]
    for i in ast.literal_eval(obj): # convertir cadena de texto a lista de diccionarios
        if c!=3: # itera sobre los elementos de la lista
            l.append(i['name']) # agrega nombre actual a la lista '1'
            c=c+1 # incrementa el contador 'c'
        else:
            break
    return l
```

```python
data['cast']=data['cast'].apply(conv)
```

```python
def conver(obj):
    l=[]
    for i in ast.literal_eval(obj): # convertir cadena de texto a lista de diccionarios
        if i['job']=='Director': # verifica si el trabajo del elemnto actual es 'Director'
            l.append(i['name']) # agrega nombre del director a la lista '1'
            break
    return l
```

```python
data['crew']=data['crew'].apply(conver)
```

```python
data['overview']=data['overview'].apply(lambda x:x.split()) # dividir 'overview' en palabras individuales
```

```python
data_genres = data[["genres"]]
```

```python
data['tags']=data['overview']+data['genres']+data['keywords']+data['cast']+data['crew'] # concatena columnas en la nueva columna 'tags'
data.drop(columns=['overview','genres','keywords','cast','crew'],axis=1,inplace=True) # elimina columnas originales
data.head()
```

```python
data['tags']=data['tags'].apply(lambda x:[i.lower() for i in x]) # convertir palabras de 'tags' a minúsculas
data['tags']=data['tags'].apply(lambda x:[i.replace(" ","") for i in x]) # elimina espacios en blanco de columna 'tags'
data.head()
```

```python
data['tags']=data['tags'].apply(lambda x:" ".join(x)) # unir palabras en cada lista de 'tags' en única cadena separada
data.head()
```

```python
import nltk
nltk.download('stopwords') # descarga recursos de palabras de parada de NLTK
```

```python
!pip install autocorrect
from autocorrect import Speller
spell = Speller(fast = True) # crear autocorrector ortográfico

from nltk.stem import PorterStemmer
ps = PorterStemmer() # crear objeto para stemming

from nltk.corpus import stopwords
stop_words = set(stopwords.words('english')) # obtener lista de stopwords en iglés

!pip install contractions
import contractions
import regex as re

def cleanup(sentence):
    sentence = re.sub("[^a-zA-Z]", " ", str(sentence)) # elimina caracter que no sea una letra
    sentence = spell(sentence) # corregir ortografía rápida
    sentence = contractions.fix(sentence) # expadir contracciones
    sentence = sentence.split() # dividir oración en palabras
    sentence = [word for word in sentence if word not in stop_words] # eliminar palabras de parada
    sentence = [ps.stem(token) for token in sentence] # stemming

    return (" ".join(sentence)) # unir palabras nuevamente en cadena
```

```python
data['tags']=data['tags'].apply(cleanup) # aplicar función 'cleanup'
data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=4806, stop_words='english')
cv_matrix = cv.fit_transform(data["tags"])
genre_vectors = cv_matrix.toarray()
genre_vectors
```

```python
from sklearn.neighbors import NearestNeighbors

knn = NearestNeighbors()
knn.fit(genre_vectors)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(genre_vectors)
similarity
```

```python
def recommend(movie):
    movie_index = data[data["title"] == movie].index[0]
    distances, indices = knn.kneighbors(similarity[movie_index].reshape(1, -1), n_neighbors=20)
    indices = indices.flatten()

    # Obtener los índices de las películas recomendadas
    recommended_indices = indices[1:20]

    # Devolver los títulos
    recommended_movies = data.iloc[recommended_indices]["title"]

    # Remover la película ingresada de la lista
    recommended_movies = recommended_movies[recommended_movies != movie]

    # Ordenar por sus distancias
    recommended_movies_with_distance = []
    for index, distance in zip(recommended_indices, distances[0][1:20]):
        title = data.iloc[index]["title"]
        recommended_movies_with_distance.append((title, distance))
    recommended_movies_with_distance = sorted(recommended_movies_with_distance, key=lambda x: x[1])

    for title, distance in recommended_movies_with_distance:
        print(title)
```

```python
recommend("Avatar")
```
